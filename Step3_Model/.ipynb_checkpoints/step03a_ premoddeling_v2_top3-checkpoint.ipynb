# Def for looping through autocorrilation values

```python
def autocorr_loop(df, lag_min, lag_max):
    lag_range = list(range(lag_min, lag_max))
    results = []
    
    for l in lag_range:
        df_shift = df.value.autocorr(lag=l)
        results.append((l, df_shift))
        
    return results
```

## Create and review autocorrilation plots form Pandas

### Top Zips

```python
plt.figure(figsize=(15,5))
figure = pd.plotting.autocorrelation_plot(top_zips.value);
```

```python
top_lags = pd.DataFrame(columns = ['lag_val', 'p_val'], data = autocorr_loop(top_zips, 1, 100))
top_lags.loc[(top_lags.p_val >= 0.95)]
```
