```python
zill = pd.read_csv('../data/processed/zillow_long_form.csv', index_col=0)
```

```python
zill.Date = pd.to_datetime(zill.Date)
```

```python
zill = zill.set_index('Date')
```

```python
zill.columns
```

```python
zill
```

```python
zill_sorted = zill.sort_values(by=['RegionName', 'Date'])
```

```python
zill_sorted['imp_val'] = zill_sorted['Value'].interpolate(method='time')
```

```python
zill_sorted
```

```python
zill_sorted.imp_val.isnull().nunique()
```

```python
zill_sorted.to_csv('../data/processed/zillow_no_nulls.csv')
```
