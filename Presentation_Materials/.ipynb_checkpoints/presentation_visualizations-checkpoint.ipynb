```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('https://mod4finalproject.s3.amazonaws.com/zillow_cleaned_no_nulls.csv',
                   index_col=0,
                   dtype = {'zip':'object'})
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df = df.drop(columns = ['regionid', 'city', 'state', 'metro', 'countyname', 'sizerank'])
```

## Lehigh, FL 33974

```python
lehigh74 = df.loc[(df.zip == '33974') & (df.index >= '2011')]
```

```python
lehigh74;
```

```python
sns.set(rc={'figure.figsize':(15,5)}, style='darkgrid', font_scale=1.5)
lehigh74_p = sns.lineplot(x=lehigh74.index, y=lehigh74.value)

lehigh74_p.set_title('Lehigh, FL 33974 Real Estate Value over Time')

lehigh74_p.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))
lehigh74_p.set(xlabel='Date', ylabel='Zillow Home Value');

fig = lehigh74_p.get_figure()
fig.savefig("lehigh74_trend.png")
```
